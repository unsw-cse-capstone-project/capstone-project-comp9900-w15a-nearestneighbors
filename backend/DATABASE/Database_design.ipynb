```python
import pandas as pd
```

```python
import sqlite3
conn = sqlite3.connect('MovieFinder.db')
c= conn.cursor()


c.execute("""CREATE TABLE Movie(
                mid integer PRIMARY KEY,
                name text,
                description text,
                region text,
                released_date text,
                director_id integer,
                votecount integer,
                rating integer,
                FOREIGN KEY (votecount) REFERENCES review(votecount),
                FOREIGN KEY (rating) REFERENCES review(rating),
                FOREIGN KEY (director_id) REFERENCES person (pid)
                )""")
c.execute("""CREATE TABLE user(
                uid integer PRIMARY KEY,
                name text,
                password text,
                profile_photo BLOB
                )""")
c.execute("""CREATE TABLE person(
                pid integer PRIMARY KEY,
                name text
                )""")


c.execute("""CREATE TABLE cast(
                castID integer,
                movieID integer,
                FOREIGN KEY (castID) REFERENCES person (pid),
                FOREIGN KEY (movieID) REFERENCES Movie (mid)
                )""")


c.execute("""CREATE TABLE review(
                userID integer,
                movieID integer,
                review_comment text,
                rating integer,
                date text,
                votecount integer,
                FOREIGN KEY (userID) REFERENCES user (uid),
                FOREIGN KEY (movieID) REFERENCES Movie (mid)
                )""")
c.execute("""CREATE TABLE watch_history(
                userID integer,
                movieID integer,
                watch_date,
                FOREIGN KEY (userID) REFERENCES user (uid),
                FOREIGN KEY (movieID) REFERENCES Movie (mid)
                )""")
c.execute("""CREATE TABLE wish_list(
                userID integer,
                movieID integer,
                FOREIGN KEY (userID) REFERENCES user (uid),
                FOREIGN KEY (movieID) REFERENCES Movie (mid)
                )""")
c.execute("""CREATE TABLE movie_genre(
                movieID integer,
                genre_type text,
                FOREIGN KEY (movieID) REFERENCES Movie (mid)
                )""")
c.execute("""CREATE TABLE user_banned_list(
                userID integer,
                banned_user_ID integer,
                FOREIGN KEY (userID) REFERENCES user (uid)
                )""")


c.execute("""CREATE TABLE Movie_Posters(
                movieID integer,
                poster BLOB,
                FOREIGN KEY (movieID) REFERENCES Movie (mid)
                )""")


conn.commit()
conn.close()
```

```python
def insertintoMoviesTable(mid,name,description,region,released_date,director_id,votecount,rating):
    sqliteConnection = sqlite3.connect('MovieFinder.db')
    cursor = sqliteConnection.cursor()
    sqlite_insert_with_param = """INSERT INTO Movie
        (mid,name,description,region,released_date,director_id,votecount,rating) 
        VALUES (?, ?, ?, ?, ?,?,?,?);""" 
    data_tuple = (mid,name,description,region,released_date,director_id,votecount,rating)
    cursor.execute(sqlite_insert_with_param, data_tuple)       
    sqliteConnection.commit()      
    cursor.close()
    
def convertToBinaryData(filename):
    with open(filename, 'rb') as file:
        blobData = file.read()
    return blobData
    
    
def insertintoPersonTable(pid,name):
    sqliteConnection = sqlite3.connect('MovieFinder.db')
    cursor = sqliteConnection.cursor()
    sqlite_insert_with_param = """INSERT INTO person
        (pid,name) 
        VALUES (?, ?);""" 
    data_tuple = (pid,name)
    cursor.execute(sqlite_insert_with_param, data_tuple)       
    sqliteConnection.commit()      
    cursor.close()
    
    
def insertintocast(castID , movieID):
    sqliteConnection = sqlite3.connect('MovieFinder.db')
    cursor = sqliteConnection.cursor()
    sqlite_insert_with_param = """INSERT INTO cast
        (castID , movieID) 
        VALUES (?, ?);""" 
    data_tuple = (castID , movieID)
    cursor.execute(sqlite_insert_with_param, data_tuple)       
    sqliteConnection.commit()      
    cursor.close()
    
def Insertintoreview( userID ,movieID ,review_comment,rating,date,votecount):
    sqliteConnection = sqlite3.connect('MovieFinder.db')
    cursor = sqliteConnection.cursor()
    sqlite_insert_with_param = """INSERT INTO review
        (userID ,movieID ,review_comment,rating,date,votecount) 
        VALUES (?, ?,?,?,?,?);""" 
    data_tuple = (userID ,movieID ,review_comment,rating,date,votecount)
    cursor.execute(sqlite_insert_with_param, data_tuple)       
    sqliteConnection.commit()      
    cursor.close()
    
def insertintowatch_history(userID , movieID ,watch_date):
    sqliteConnection = sqlite3.connect('MovieFinder.db')
    cursor = sqliteConnection.cursor()
    sqlite_insert_with_param = """INSERT INTO watch_history
        (userID , movieID ,watch_date) 
        VALUES (?, ?,?);""" 
    data_tuple = (userID , movieID ,watch_date)
    cursor.execute(sqlite_insert_with_param, data_tuple)       
    sqliteConnection.commit()      
    cursor.close()   
def insertintowish_list(userID , movieID):
    sqliteConnection = sqlite3.connect('MovieFinder.db')
    cursor = sqliteConnection.cursor()
    sqlite_insert_with_param = """INSERT INTO wish_list
        (userID , movieID) 
        VALUES (?, ?);""" 
    data_tuple = (userID , movieID )
    cursor.execute(sqlite_insert_with_param, data_tuple)       
    sqliteConnection.commit()      
    cursor.close()      
def insertintomovie_genre(movieID ,genre_type):
    sqliteConnection = sqlite3.connect('MovieFinder.db')
    cursor = sqliteConnection.cursor()
    sqlite_insert_with_param = """INSERT INTO movie_genre
        (movieID ,genre_type)
        VALUES (?, ?);""" 
    data_tuple = (movieID ,genre_type )
    cursor.execute(sqlite_insert_with_param, data_tuple)       
    sqliteConnection.commit()      
    cursor.close()      
def insertintouser_banned_list( userID,banned_user_ID):
    sqliteConnection = sqlite3.connect('MovieFinder.db')
    cursor = sqliteConnection.cursor()
    sqlite_insert_with_param = """INSERT INTO user_banned_list
        ( userID,banned_user_ID)
        VALUES (?, ?);""" 
    data_tuple = (userID,banned_user_ID)
    cursor.execute(sqlite_insert_with_param, data_tuple)       
    sqliteConnection.commit()      
    cursor.close()     
def convertToBinaryData(filename):
    with open(filename, 'rb') as file:
        blobData = file.read()
    return blobData


def insertintoUserTable(uid,name,password,profile_photo):
    sqliteConnection = sqlite3.connect('MovieFinder.db')
    cursor = sqliteConnection.cursor()
    sqlite_insert_with_param = """INSERT INTO user
        (uid,name,password,profile_photo) 
        VALUES (?, ?, ?, ?);""" 
    Photo = convertToBinaryData(profile_photo)
    data_tuple = (uid,name,password,Photo)
    cursor.execute(sqlite_insert_with_param, data_tuple)       
    sqliteConnection.commit()      
    cursor.close()
def insertintopostersTable(movieID,poster):
    sqliteConnection = sqlite3.connect('MovieFinder.db')
    cursor = sqliteConnection.cursor()
    sqlite_insert_with_param = """INSERT INTO Movie_Posters
        (movieID,poster) 
        VALUES (?, ?);""" 
    Photo = convertToBinaryData(poster)
    data_tuple = (movieID,Photo)
    cursor.execute(sqlite_insert_with_param, data_tuple)       
    sqliteConnection.commit()      
    cursor.close()
```
