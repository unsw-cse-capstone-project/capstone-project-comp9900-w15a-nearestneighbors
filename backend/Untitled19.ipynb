```
import pandas as pd
df = pd.read_csv("tmdb_5000_movies.csv")
df1= pd.read_csv("tmdb_5000_credits.csv")
df.columns.values[3] = 'movie_id'
print(df.columns)
print(df1.columns)
movies = df.merge(df1,on = 'movie_id')
```

```
movies.head()
```

```
def get_genre(row):
  try: 
    return eval(row)[0]['name']
  except IndexError:
    return 'Unknown'
movies['genre_final'] = movies['genres'].apply(get_genre)
```

```
def get_region(row):
    try: 
      return eval(row)[0]['name']
    except IndexError:
      return 'Unknown'

movies['region_final'] = movies['production_countries'].apply(get_region)
```

```
def get_cast(row):
  lister = []
  try:
    
    for i in eval(row):
      lister.append(i['character'])
    return lister
  except IndexError:
    return lister


movies['cast_final'] = movies['cast'].apply(get_cast)
```

```
movies.head()
```

```
movies_imp = movies[['original_title', 'overview','region_final', 'release_date', 'genre_final', 'cast_final', 'director']]
```

```
movies_imp.head()
```

```
movies_imp.to_csv('movies_final.csv')
```
